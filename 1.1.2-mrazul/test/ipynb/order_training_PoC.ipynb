```python
import h2o
from pandas import DataFrame as DataFrame
from pandas import concat as concat
import json
import numpy as np
```

```python
import os
import sys
import tarfile
from six.moves import cPickle as pickle
# In[2]:

os.chdir('d:/Data/Gdeeplearning-Udacity')
pickle_file = 'notMNIST.pickle'

with open(pickle_file, 'rb') as f:
  save = pickle.load(f)
  train_dataset = save['valid_dataset']
  train_labels = save['valid_labels']
  valid_dataset = save['test_dataset']
  valid_labels = save['test_labels']

  del save  # hint to help gc free up memory
    
print('Training set', train_dataset.shape, train_labels.shape)
print('Validation set', valid_dataset.shape, valid_labels.shape)

h2o.init(nthreads=6)
```

```python
# In[5]:

image_size = 28
num_labels = 10

def reformat(dataset, labels):
    dataset = DataFrame(dataset.reshape((-1, image_size * image_size)).astype(np.float32))
    # Map 0 to [1.0, 0.0, 0.0 ...], 1 to [0.0, 1.0, 0.0 ...]
    columns = list()
    for each_value in dataset.columns.values:
        columns.append(str(each_value))
    dataset.columns = columns
    labels = DataFrame(labels.reshape((-1, 1)).astype(str))
    columns = list()
    for each_value in labels.columns.values:
        columns.append('objective'+ str(each_value))
    labels.columns = columns
    #labels = (np.arange(num_labels) == labels[:,None]).astype(np.float32)
    return dataset, labels

pd_train_dataset, pd_train_labels = reformat(train_dataset, train_labels)
pd_valid_dataset, pd_valid_labels = reformat(valid_dataset, valid_labels)
pd_train_dataset = concat([pd_train_dataset,pd_train_labels],axis=1)
pd_valid_dataset = concat([pd_valid_dataset,pd_valid_labels],axis=1)
                       
print('Training set', pd_train_dataset.shape)
print('Validation set', pd_valid_dataset.shape)

# In[9]:

#deleting non used memory
#train_dataset = []
#train_labels = []
#valid_dataset = []
#valid_labels = []
```

```python
from h2o import H2OFrame as H2OFrame
import time

def order_training(analysis_id, training_frame, valid_frame, analysis_list):
    assert isinstance(analysis_id, str)
    assert isinstance(training_frame, DataFrame)
    assert isinstance(valid_frame, DataFrame) or valid_frame is None
    assert isinstance(analysis_list, list)

    _analysis_id = analysis_id  # "customer_" + #random + #timestamp
    _status = -1  # Operation Code
    _analysis_list = analysis_list  # [(ar(incomplete).json, normalizations_sets.json)]
    _model_list = list()
    
    _training_frame = h2o.H2OFrame(python_obj=pd_train_dataset)
    if valid_frame is not None:
        _valid_frame = h2o.H2OFrame(python_obj=pd_valid_dataset)

    for algorithm_description, normalization in _analysis_list:
        # Mapping
        print(algorithm_description)
        print(normalization)
        struct_ar = json.load(algorithm_description)
                                                                       
        if normalization is not None:
            struct_ns = json.load(algorithm_description)
        else:
            struct_ns = None
                                                                       
        assert isinstance(struct_ar, dict)
        assert isinstance(struct_ns, dict) or normalization is None
        
        _model_id = struct_ar['model_id']                                                                     
        y = struct_ar['model_parameters']['h2o'][0]['parameters']['response_column']['value']
        print(y)
        x = _training_frame.col_names
        print(x)
        x.remove(y)  
                                                                       
        for each_type in struct_ar['model_parameters']['h2o'][0]['types']:
            if each_type['active']:
                if  each_type['type'] in ['binomial','multinomial']: 
                    _training_frame[y] = _training_frame[y].asfactor()
                    if valid_frame is not None:
                        _valid_frame[y] = _valid_frame[y].asfactor()
                else: 
                    _training_frame[y] = _training_frame[y].asfloat()
                    if valid_frame is not None:
                        _valid_frame[y] = _valid_frame[y].asfloat()                   
                    
        model_command = list()
        model_command.append(struct_ar['model_parameters']['h2o'][0]['module'])
        model_command.append(".")
        model_command.append(struct_ar['model_parameters']['h2o'][0]['model'])
        model_command.append("(")
        # model_command.append("model_id=\'%s\'"% struct_ar['model_id'])
        model_command.append("training_frame=_training_frame")                                                                     
        if valid_frame is not None:
            model_command.append(", validation_frame=_valid_frame")
                                                                       
        for key, value in struct_ar['model_parameters']['h2o'][0]['parameters'].items():
            if value['seleccionable']:
                if isinstance(value['value'], str):
                    model_command.append(", %s=\'%s\'" % (key, value['value']))
                else:
                    model_command.append(", %s=%s" % (key, value['value']))
                                                                       
        model_command.append(")")
        model_command = ''.join(model_command)
        print(model_command)
        model_base = eval(model_command)

        if valid_frame is not None:
            model_base.train(x=x, y=y, training_frame=_training_frame, validation_frame=_valid_frame)
        else:
            model_base.train(x=x, y=y, training_frame=_training_frame)
        
        load_path = list()
        load_path.append(struct_ar['load_path']['value'])
        load_path.append('/')
        load_path.append(struct_ar['model_id'])
        load_path.append('/')
        load_path.append(str(time.time()))
        load_path.append('.h20')
        
        h2o.save_model(model=model_base, path=''.join(load_path), force=True)
        self._model_list.append(model_base)

    return _model_id, _model_list.copy()
```
